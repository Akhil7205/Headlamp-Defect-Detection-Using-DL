```python

from PIL import Image
def load_images_from_folder(folder, label, img_size=(64,64)):
    images, labels = [], []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        try:
            img = Image.open(img_path).convert("RGB")
            img = img.resize(img_size)
            images.append(np.array(img))
            labels.append(label)
        except:
            continue
    return images, labels
```

```python
intact_images, intact_labels = load_images_from_folder("data/intact_img", 0)
broken_images, broken_labels = load_images_from_folder("data/broken_img", 1)
moisture_images, moisture_labels = load_images_from_folder("data/moisture_img", 2)

X = np.array(intact_images + broken_images + moisture_images)
y = np.array(intact_labels + broken_labels + moisture_labels)

```

```python
X[50]
```

```python
# Normalize
X = X / 255.0
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode="nearest"
)
datagen.fit(X_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Flatten(input_shape=(64,64,3)),
    Dense(512, activation="relu"),
    BatchNormalization(),
    Dropout(0.4),

    Dense(256, activation="relu"),
    BatchNormalization(),
    Dropout(0.3),

    Dense(128, activation="relu"),
    Dense(3, activation="softmax")
])

model.compile(
    optimizer=Adam(0.0005),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history = model.fit(
    datagen.flow(X_train, y_train, batch_size=8), validation_data=(X_test, y_test),epochs=20
)
```

```python
y_pred = np.argmax(model.predict(X_test), axis=1)
print("Unique predictions:", np.unique(y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import json
from tensorflow.keras.preprocessing import image
import numpy as np

img_path = "/content/data/broken_img/img1.jpg"
img = image.load_img(img_path, target_size=(64, 64))
img_array = image.img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# Predict
prediction = model.predict(img_array)
predicted_class = np.argmax(prediction, axis=1)[0]

class_names = ["Intact", "Broken", "Moisture"]
output = {
    "predicted_class": class_names[predicted_class],
    "probabilities": {
        class_names[i]: float(prediction[0][i]) for i in range(len(class_names))
    }
}

# Print JSON result
print(json.dumps(output, indent=4))
```
